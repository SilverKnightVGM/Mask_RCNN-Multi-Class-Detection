```python
import json
from tqdm import tqdm

with open('test.json') as json_file:
    data = json.load(json_file)
```

```python
image_names = list(data.keys())
n_images = len(image_names);
print(n_images)
print(image_names[:5])
```

```python
img = image_names[0]
data[img]
```

```python
n_regions = data[img].count('cibo')
a = data[img]
# print(*a[-4:-2], sep="\n\n")
print(*a[0::n_regions], sep="\n\n")
print("----------------")
print(*a[1::n_regions], sep="\n\n")
```

```python
# Create List of all classes (73) and mappings to integers
all_classes = []
for image_name in image_names:
    n_regions = data[image_name].count('cibo')
#     print("Name:{}".format(image_name))
#     print("#Regions:{}".format(n_regions))
    for i in range(n_regions):
        # Each row is one region/item, i is region number
        row = data[image_name][i::n_regions]
        # Extract all classes
        i_class = row[1]
#         i_name = row[2]
        all_classes.append(i_class)

all_classes = list(set(all_classes))
all_classes.sort()
# print("{} classes found!".format(len(all_classes)))
# print(all_classes)
# print(all_classes.index("budino"))

class_mappings = {}
for i,c in enumerate(all_classes):
    class_mappings[c] = str(i)

print(class_mappings)
class_mappings['budino']
```

```python
master_dict = {}

for image_name in tqdm(image_names):
    n_regions = data[image_name].count('cibo')
    regions = {}
#     print("Name:{}".format(image_name))
#     print("#Regions:{}".format(n_regions))
#     print("\n")
    for i in range(n_regions):
        shape_attributes = {}
        region_attributes = {}
        # Each row is one region/item, i is region number
        row = data[image_name][i::n_regions]
        # Extract all info
        i_category = row[0]
        i_class = row[1]
        i_name = row[2]
        
        i_boundType = row[3]
        if(i_boundType[:-1] != "Polygon"): 
            print("non polygon check")

        i_points = row[4]
        i_all_x = i_points[0::2]
        i_all_y = i_points[1::2]

        i_bbox = row[5]
        
        # Create regions
        shape_attributes["name"] = i_boundType[:-1]
        shape_attributes["all_points_x"] = i_all_x
        shape_attributes["all_points_y"] = i_all_y
        region_attributes["food"] = class_mappings[i_class]
        
        regions[i] = {"shape_attributes":shape_attributes, "region_attributes": region_attributes}
    
    master_dict[image_name+".jpg"] = {"filename":image_name+".jpg","regions":regions}
```

```python
print(len(master_dict.keys()))
master_dict.keys()
```

```python
with open('annotations.json', 'w') as fp:
    json.dump(master_dict, fp, indent=2)
print("Done")
```
